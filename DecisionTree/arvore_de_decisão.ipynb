```
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

import pandas as pd

# url = "https://raw.githubusercontent.com/JoaoPauloFialho/machine-learning/main/datasets/mushroomshabitat.data"
url = "https://raw.githubusercontent.com/JoaoPauloFialho/machine-learning/main/datasets/wineorigin.data"

# Carregar base de dados
# DataFrame
dataset = pd.read_csv(url, header=None)

columns = len(dataset.columns)

# y = dataset[collumns-1] # extrai a primeira coluna, que é o habitat do cogumelo
y = dataset[0] # extrai a primeira coluna, que é a origem do vinho
X = dataset.loc[:,1:columns-1]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) # 80% treino e 20% teste
#stratify=y é para manter a proporção da classe y
#random_state=None é para que sempre que o test rode ele embaralha de maneira aleatória, se o valor de random_state fosse 42 por exemplo, o embaralhamento do teste sempre seria o mesmo
```

```
# Treinamendo da Árvore de Decisão
#model = tree.DecisionTreeClassifier(criterion="entropy",max_depth=2)
model = tree.DecisionTreeClassifier(criterion="entropy") #o padrão é gini
model = model.fit(X_train, y_train)
```

```
# Predição e Resultados

result = model.predict(X_test)
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

print(list(result))
print(list(y_test))
```
